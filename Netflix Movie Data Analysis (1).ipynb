**Categorizing 'Vote_Average' column** 

I would cut the Vote_Average values and make 4 categories: 'Popular' 'Average' 'Below_Avg' 'Low_Avg' to describe it more using 'Categorize_col()' function
provided above.

```python
def catigorize_col(df, col, labels):
    edges = [
        df[col].describe()['min'],
        df[col].describe()['25%'],
        df[col].describe()['50%'],
        df[col].describe()['75%'],
        df[col].describe()['max']
    ]
    df[col] = pd.cut(df[col], edges, labels=labels, duplicates='drop')
    return df
```

```python
labels = ['Not_popular', 'Below_avg', 'Average', 'Popular']

df = catigorize_col(df, 'Vote_Average', labels)

df['Vote_Average'].unique()
```
